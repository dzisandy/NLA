1. Let's consider $ Y = AXB$ Then:
$$
Y_{ij} = (AXB)_{ij} = (AX)_i B^T_{j} = A_i X B^T_j
$$
Now, let's consider $(B^T \otimes A)\, \mathrm{vec}(X)$: 
$$
Z_{j*n + i} = ((B^T \otimes A)\, \mathrm{vec}(X))_{j*n + i} = \sum_{k} B_{kj} (A_i X^T_{k}) = B^T_j A_i X = A_i X B^T_j
$$
Where $M_i$ - $i$-th row of matrix M. So, one can see that $Y_{ij} = Z_{j*n + i} \Rightarrow \mathrm{vec}(AXB) = (B^T \otimes A)\, \mathrm{vec}(X)$ . For row wise reshape, we have the same problem $(AXB)^T = B^T X^T A^T$. Due to the fact that $\mathrm{vec}_{row}(X^T) = \mathrm{vec}_{column}(X)$ we will have: 
$$
\mathrm{vec}_{row}(AXB) = \mathrm{vec}_{column}((AXB)^T) = \mathrm{vec}_{column}(B^T X^T A^T) = (A \otimes B^T)\, \mathrm{vec}_{column}(X^T) = (A \otimes B^T)\, \mathrm{vec}_{row}(X)
$$


2. Naive multiplication for case $(A \otimes B) x$ is $O(n^4)$, beacuse it is $n^2 \times n^2$ matrix mult $n^2$ vector, and for matvec operations we have $O(l^2)$, where $ l $ is size of vector. It can be reduced by equation $\mathrm{vec}(AXB)=(A \otimes B^\top) \mathrm{vec}(X).$ and multiplication will have difficulty $O(n^3)$ or maybe less if we will use not naive matrix multiplication (ex. strassen).

3. Let's find eigen values, using definition:
$$
(A\otimes I + I \otimes B)vec(X) = \lambda vec(X) \\
vec(A X) + vec(X B^T) = \lambda vec(X)\\
A X + X B^T = \lambda X
$$

Here used previous formula. Now we need to find $\lambda$ from this equation. Let's use eigendecomposition $A = S_A\Lambda_A S_A^{-1}$ and $B = S_B\Lambda_B S^{-1}_B$: 
$$
S_A\Lambda_A S_A^{-1} X + X (S^T)^{-1}_B \Lambda_B S_B^T = \lambda X\\
\Lambda_A S_A^{-1} X (S^T)^{-1}_B + S_A^{-1} X (S^T)^{-1}_B\Lambda_B = S_A^{-1} \lambda X (S^T)^{-1}_B \\
$$

Now , use new variable $Y = S_A^{-1} X (S^T)^{-1}_B$:

$$
\Lambda_A Y + Y \Lambda_B = \lambda Y\\
(\Lambda_A - \lambda I) Y + Y \Lambda_B = 0 
$$

Now let's consider vectors again:
$$
((\Lambda_A - \lambda I) \otimes I + I \otimes \Lambda_B) vec(Y) = 0 
$$


Now to find eigenvalues one need to understand when $det(Z(\lambda)) = 0$ , where $Z(\lambda) =(\Lambda_A - \lambda I) \otimes I + I \otimes \Lambda_B $. Using the fact that $Z$ is diagonal matrix, we have that eigenvectors $vec(Y_k) = e_k$, where $e_k$ is vector with one in $k$-th position,others are zeros and for eigenvalues we have:
$$
det(Z(\lambda)) = \prod_{i,j}((\Lambda_A)_{ii} + (\Lambda_B)_{jj} - \lambda) = 0
$$
So, one can see that eigenvalues of matrix $A\otimes I + I \otimes B$ is $diag(\Lambda_A \otimes I + I \otimes \Lambda_B)$ . So, for eigenvectors of $A\otimes I + I \otimes B$:
$$
vec(Y) = vec(S_A^{-1} X (S^T)^{-1}_B) \Rightarrow vec(X_k) = vec(S_A Y_k S_B^T) 
$$
4. Here we in next three tasks we used facts from lectures about convergence. Richardson iteration: We have $A = A^*$ and $A > 0$, so the optimal parameter will be $\tau_{opt} = \frac{2}{\lambda_{max} + \lambda_{min}} = 0.001998$ and $q_{opt} = \frac{\lambda_{max} -\lambda_{min}}{\lambda_{max} + \lambda_{min}} = \frac{999.999}{1000.001} = 0.999998 $. So, number of iterations: $k = \log_q(10^{-4}) = -4\log_q 10 = 4605000$

5. Chebyshev iteration: We have $A = A^*$ and $A > 0$, so the optimal parameter for every step will be: 
$$
\tau_i = \frac{2}{\lambda_\max + \lambda_\min - (\lambda_\max - \lambda_\min)x_i}, \quad x_i = \cos \frac{\pi(2i + 1)}{2n}\quad i=0,...,n-1
$$
\\
And for convergence speed we will have $q =  \frac{\sqrt{\mathrm{cond}(A)}-1}{\sqrt{\mathrm{cond}(A)}+1} = \frac{\sqrt{\frac{\lambda_\max}{\lambda_\min}}-1}{\sqrt{\frac{\lambda_\max}{\lambda_\min}}+1} = \frac{999}{1001} = 0.998 $. ## Problem 2 (40 pts)

### Spectral graph partitioning and inverse iteration


Given connected graph $G$ and its corresponding graph Laplacian matrix $L = D - A$ with eigenvalues $0=\lambda_1, \lambda_2, ..., \lambda_n$, where $D$ is its degree matrix and $A$ is its adjacency matrix, *Fiedler vector* is an eignevector correspondng to the second smallest eigenvalue $\lambda_2$ of $L$. Fiedler vector can be used for graph partitioning: positive values correspond to the one part of a graph and negative values to another.

### Inverse power method (15 pts)

To find the Fiedler vector we will use the inverse iteration with adaptive shifts (Rayleigh quotient iteration). 

* (5 pts) Write down the orthoprojection matrix on the space orthogonal to the eigenvector of $L$, corresponding to the eigenvalue $0$ and prove (analytically) that it is indeed an orthoprojection.
 
* (5 pts) Implement the spectral partitioning as the function ```partition```:

The projection of any vector $x$ on the space orthogonal to the first eigenvector of $L$ is:
$$
proj(x) = x - \frac{(x;eig) eig}{\| eig\|_2^2}
$$
So, the  left projection matrix $A$ will be:
$$
A = I - diag((eig_1, eig_2,..., eig_n))\frac{E}{\| eig\|_2^2}
$$
Where E - matrix, which consist from all ones. In our case, eigenvector is $(1,1,1,....,1,1,1)$. So in the end we have:
$$
A = I - \frac{E}{n}
$$
Proof: Let $x$ be any vector. Then:
$$
(Ax;eig) = (x;eig) - (x;\frac{E}{n}eig) = (x;eig) - (x;eig) = 0
$$
So, it is really projection matrix.

````python
# INPUT:
# A - adjacency matrix (scipy.sparse.csr_matrix)
# num_iter_fix - number of iterations with fixed shift (int)
# shift - (float number)
# num_iter_adapt - number of iterations with adaptive shift (int) -- Rayleigh quotient iteration steps
# x0 - initial guess (1D numpy.ndarray)
# OUTPUT:
# x - normalized Fiedler vector (1D numpy.ndarray)
# eigs - eigenvalue estimations at each step (1D numpy.ndarray)
# eps - relative tolerance (float)
import scipy.sparse as sp
import scipy.sparse.linalg
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import networkx as nx



def create_lapl(A):
    D = sp.dia_matrix( (A.sum(1).T,[0]), shape = A.shape)
    return D - A


def find_proj(x):
    mat_proj = x - x.sum() * (1/x.shape[0])
    return mat_proj



def partition(A, shift, num_iter_fix, num_iter_adapt, x0, eps):
    x = x0
    x = find_proj(x/np.linalg.norm(x))
    eigs = []
    L = create_lapl(A)
    Lambda_1 = -1
    epsilon = []
    for i in range(num_iter_fix):
        Lambda = x.dot(L.dot(x))
        x = sp.linalg.spsolve( (L - shift * sp.identity(A.shape[0])) ,x)
        x = find_proj(x/np.linalg.norm(x))
        epsilon.append(np.abs(Lambda - Lambda_1)/np.abs(Lambda))
        if epsilon[-1] < eps:
            break
        Lambda_1 = Lambda
        eigs.append(Lambda_1)

    
    for i in range(num_iter_adapt):

        Lambda = x.dot(L.dot(x))
        x = sp.linalg.spsolve( (L - Lambda * sp.identity(A.shape[0])) ,x)
        x = x/np.linalg.norm(x)
        #x = find_proj(x/np.linalg.norm(x))
        
        epsilon.append(np.abs(Lambda - Lambda_1)/np.abs(Lambda))
        if epsilon[-1] < eps:
            break
        Lambda_1 = Lambda
        eigs.append(Lambda_1)
       
    return x, eigs,epsilon
````

Algorithm must halt before `num_iter_fix + num_iter_adapt` iterations if the following condition is satisfied $$ \boxed{\|\lambda_k - \lambda_{k-1}\|_2 / \|\lambda_k\|_2 \leq \varepsilon} \text{ at some step } k.$$

Do not forget to use the orthogonal projection from above in the iterative process to get the correct eigenvector.
It is also a good idea to use ```shift=0``` before the adaptive stragy is used. This, however, is not possible since the matrix $L$ is singular, and sparse decompositions in ```scipy``` does not work in this case. Therefore, we first use a very small shift instead.

Let graph with Laplacian matrix $L$ have $k$ connected components. Then Laplacian matrix will be block matrix:
$$
L = \begin{pmatrix}
L_1 & 0 & ... 0\\
 0 & L_2 & ... 0 \\
0 & 0 & ... L_k 
\end{pmatrix}
$$
Where $L_i$ - Laplacian matrix of $k$-th graph component separetly. Any $L_k$ is Laplacian matrix of 1-component graph and has one eigenvector corresponding to zero eigenvalue. So, Let $L_k \in \mathbf{R}^{n_k \times n_k}$ we have :
$$
Lx = 0 \Leftrightarrow L_j x[\sum_{i=1}^{j-1}n_i:\sum_{i=1}^{j-1}n_i + n_k] = 0 \ for \ any \ j \in \{1,2,...,k\}
$$
Any of this equations have only trivial solution and $x[\sum_{i=1}^{j-1}n_i:\sum_{i=1}^{j-1}n_i + n_k] = C(1,1,1,...,1,1)$.
Let's consider  any $x$ from subspace $A = \{x \in \mathbf{R}^n | x =\sum_{i=1}^{k} \alpha_i x_i, \alpha_i \in \mathbf{R}\}$, where $x_i[\sum_{j=1}^{i-1}n_i:\sum_{j=1}^{i-1}n_i + n_k] = (1,1,1,...,1,1)$ and other elements of $x_i$ is zeros. This vector satisfy the equations for $L_k$, so they are solutions of the system $Lx = 0$. Vector which are not in this subspace doen't satisfy the equations for $L_k$, so they are not solution.

So, subspace $A = L^{\perp}$, which defined by linear system $Lx = 0$ will have dimention $dim(L^{\perp}) = k$. Thus, multiplicity of eigenvector 0 equals k.

Let us deal here with a graph constructed from a binarized image.

* (5 pts) Write a function that constructs sparse adjacency matrix from the binarized image, taking into account the rule from above.

````python


%matplotlib inline
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw
import requests
import numpy as np



def create_adj(image):
    d = {}
    counter = 0
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            
            if image[i,j] == 0:
                if counter == 0:
                    first_key = str(i) + "," + str(j)
                d[str(i) + "," + str(j)] = counter
                counter += 1
    
    
    key = first_key            
    i = int(key.split(",")[0])
    j = int(key.split(",")[1])
    row_res = np.zeros(len(d))
    
    for s_i in range(3):
        for s_j in range(3):
            if s_i == 1 and s_j == 1:
                continue
            if i + s_i - 1 >= 0 and j + s_j - 1 >= 0 and \
            i + s_i - 1 < image.shape[0] and j + s_j - 1 < image.shape[1]:
                if image[i + s_i - 1][j + s_j - 1] == 0:
                    row_res[d[str(i + s_i - 1) + ',' + str(j + s_j - 1)]] = 1
    
    row_res = sp.csr_matrix(row_res)
                        
                    
    for index,key in enumerate(d.keys()):
        if index == 0:
            continue
        i = int(key.split(",")[0])
        j = int(key.split(",")[1])
        row = np.zeros(len(d))
        for s_i in range(3):
            for s_j in range(3):
                if s_i == 1 and s_j == 1:
                    continue
                if i + s_i - 1 >= 0 and j + s_j - 1 >= 0 and \
                i + s_i - 1 < image.shape[0] and j + s_j - 1 < image.shape[1]:
                    if image[i + s_i - 1][j + s_j - 1] == 0:
                        row[ d[str(i + s_i - 1) + ',' + str(j + s_j - 1)] ] = 1
                     
        row_res = sp.vstack((row_res, sp.csr_matrix(row)))
    return d,row_res



def get_bin_image(image):
    coef = 30
    return np.maximum(np.sign((coef - image)), 0)

url = 'http://chittagongit.com//images/github-icon-svg/github-icon-svg-16.jpg'
image_raw = Image.open(requests.get(url, stream=True).raw)
image_raw = image_raw.convert('L')
image = np.array(image_raw).astype(np.float)

image = get_bin_image(image)






plt.imshow(Image.fromarray(image*255))
plt.xticks(())
plt.yticks(())
plt.title('Original Picture')
plt.show()

d,adj_mat = create_adj(image)


````

### 1. Kernel (5 pts)

````python
import scipy.special
````

````python

k0 = 50#
N = 300#




def make_eG(k0, N):
    # INPUT:  
    # k0 #dtype = float
    # N #dtype = int
    
    x,y = np.meshgrid(np.fft.fftfreq(2*N - 1,1/(2*N - 1)).astype(np.complex64),
                                np.fft.fftfreq(2*N - 1,1/(2*N - 1)).astype(np.complex64))
    

    dr = np.sqrt(x*x + y*y)/(N-1)
    
    
    eG = complex(0,-1/4) * scipy.special.hankel2(0,k0*dr)
    eG[0,0] = 0
    

    # OUTPUT:
    # np.array, shape = (2N-1, 2N-1), dtype = np.complex64
    return eG



eG = make_eG(k0=k0, N=N)

plt.imshow(eG.real)
````

### 2. Matvec (5 pts)

````python
def Gx(x,eG):
    # input:  
    # x, np.array, shape=(N**2, ), dtype = np.complex64
    # eG, np.array, shape=(2N-1, 2N-1), dtype = np.complex64
    # output:
    # matvec, np.array, shape = (N**2, ), dtype = np.complex64
    matvec = np.fft.ifft2(np.fft.fft2(eG) * np.fft.fft2( np.block([[x.reshape(N,N),np.zeros(N,N-1)],
                                                                    [np.zeros(N-1,2*N - 1)]] )))
    return matvec[0:N, 0:N].reshape(N**2,)
````

### 3. Complexity (3 pts)

Big-O complexity of one matvec operation is $O(M^2log(M))$ , where $M = 2N - 1$. Because  2-dim fft consist from 1-dim ffts applied to all rows and columns. Every operation have difficullty $O(MlogM)$ and number of operation is $2*M$,where $M$ is length of row or column. So, complexity is $O(M^2log(M))$.
