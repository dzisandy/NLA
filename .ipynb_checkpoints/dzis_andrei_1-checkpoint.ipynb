* (5 pts) Write a function 
```python
def gen_toeplitz(N, alpha):    
    return T
```
that outputs matrix $T$: $$T_{ij} = \sqrt{\frac{\alpha}{\pi}}e^{-\alpha (i-j)^2}, \quad i,j=1,\dots,N$$ as numpy array. <font color='red'> Avoid using loops or lists! </font> The function [np.meshgrid](https://docs.scipy.org/doc/numpy-1.13.0/reference/generated/numpy.meshgrid.html) will be helpful for this task.
**Note:** matrices that depend only on difference of indices: $T_{ij} \equiv T_{i-j}$ are called **Toeplitz**. Toeplitz matrix-by-vector multiplication is **convolution** since it can be written as $$y_i = \sum_{j=1}^N T_{i-j} x_j.$$ Convolutions can be computed faster than $\mathcal{O}(N^2)$ complexity using Fast Fourier transform (will be covered later in our course, no need to implement it here).

````python
# INPUT: N - integer (positive), alpha - float (positive)
# OUTPUT: T - np.array (shape: NxN)
def gen_toeplitz(N, alpha):
    # Your code is here
    a = np.arange(1,N+1)
    b = np.arange(1,N+1)
    i,j = np.meshgrid(a,b)
    T = np.sqrt(alpha/np.pi)*np.exp(-alpha*(i-j)**2)
    return T
````

## Convolution (10 pts)

* (5 pts) Write a function ```convolution``` (see below)
that takes the signal you want to convolve and multiply it by Toeplitz matrix T (for matvec operations use @ symbol). 

````python
# INPUT: signal - np.array (shape: Nx1), N - int (positive), alpha - float (positive)
# OUTPUT: convolved_signal - np.array (shape: Nx1)

def convolution(signal, N, alpha):
    # Your code is here
    matrix = gen_toeplitz(N,alpha)
    convolved_signal = signal @ matrix
    return convolved_signal
````

## Deconvolution (20 pts)

Given a convolved signal $y$ and an initial signal $x$ our goal now is to recover $x$ by solving the system
$$
    y = Tx.
$$
To do so we will run iterative process
$$
    x_{k+1} = x_{k} - \tau_k (Tx_k - y), \quad k=1,2,\dots
$$
starting from zero vector $x_0$. There are different ways how to define parameters $\tau_k$.
Different choices lead to different methods (e.g. Richardson iteration, Chebyshev iteration, etc.).
This topic will be covered in details later in our course.

To get some intuition why this process converges to the solution of $Tx=y$, we can consider the following. Let us note that if $x_k$ converges to some limit $x$, then so does $x_{k+1}$. Taking $k\to \infty$ we arrive at $x = x - \tau (Tx -  y)$ and hence $x$ is the solution of $Tx = y$. 

Another important point is that iterative process requires only matrix-vector porducts $Tx_k$ on each iteration instead of the whole matrix. In this problem we, however, work with the full matrix, but keep in mind, that convolution can be done efficiently without storing the whole matrix.

* (5 pts) For each $k$ choose paremeter $\tau_k$ such that the residual $r_{k+1}=Tx_{k+1} - y$ is minimal possible (*line search* with search direction $r_k$):
$$
    \|Tx_{k+1} - y\|_2 \to \min_{\tau_k}
$$
found analytically. The answer to this bullet is a derivation of $\tau_k$. The parameter $\tau_k$ should be expressed in terms of residuals $r_k = T x_k - y$.

````python
# Your solution is here
def tau(r_k, T):
    return np.dot(r_k, T @ r_k)/np.dot(T @ r_k, T @ r_k)
    
````

* (10 pts) Write a function ```iterative```
that outputs accuracy –– a numpy array of relative errors $\big\{\frac{\|x_{k+1} - x\|_2}{\|x\|_2}\big\}$ after ```num_iter``` iterations using $\tau_k$ from the previous task. **Note:** The only loop you are allowed to use here is a loop for $k$.

````python
# INPUT:  N - int (positive), alpha - float (positive), num_iter - integer (positive), 
#  y - np.array (shape: Nx1, convolved signal), s - np.array (shape: Nx1, original signal)
# OUTPUT: rel_error - np.array size (num_iter x 1)

def iterative(N, num_iter, y, s, alpha):
    # Your code is here
    error = np.zeros(num_iter)
    x_k = np.zeros(y.shape)
    T = gen_toeplitz(N, alpha)
    for k in range(num_iter):
        tau_k = tau(T @ x_k - y,T)
        x_k = x_k - tau_k*(T @ x_k - y)
        error[k] = np.linalg.norm(x_k-s,2)/np.linalg.norm(s,2)
    return error
````

Now the iterative process turns into: $x_{k+1} = x_{k} - \tau_k (Tx_k - y_{noisy}) \rightarrow x_{k+1} - x_{k}  = - \tau_k (Tx_k - y - noise) \rightarrow$ if $k \rightarrow \infty $, then $\frac{\|x_{k+1} - x\|_2}{\|x\|_2} \rightarrow \infty $

So adding small noise turned into inrease of the relative error.

# Problem 3 (Strassen algorithm) 15 pts

_1._ (3 pts) Implement the naive algorithm for squared matrix multiplication with explicit “for” cycles.

````python
def naive_multiplication(A, B):
    """
    Implement naive matrix multiplication with explicit for cycles
    
    Parameters: Matrices A, B
    
    Returns: Matrix C = AB
    """
#     Your code is here 
    N = int(A.shape[0])
    C = np.zeros(shape = (N,N))
    for i in range(N):
        for j in range(N):
            for k in range(N):
                C[i][j] += A[i][k] * B[k][j]
    return C
````

_2._ (7 pts) Implement the Strassen algorithm.

````python
def strassen(A, B):
    """
    Implement Strassen algorithm for matrix multiplication 
    
    Parameters: Matrices A, B
    
    Returns: Matrix C = AB
    """
#     Your code is here
    #strassen algo works if N == 2^n, cause of splitting, for another N's use reshape(A,B)
    # formulas provided by https://ru.wikipedia.org/wiki/Алгоритм_Штрассена
    shape = A.shape[0]
    if shape ==1:
        return A*B
    A11 = A[:shape//2, :shape//2]
    A12 = A[:shape//2, shape//2:]
    A21 = A[shape//2:, :shape//2]
    A22 = A[shape//2:, shape//2:]
    B11 = B[:shape//2, :shape//2]
    B12 = B[:shape//2, shape//2:]
    B21 = B[shape//2:, :shape//2]
    B22 = B[shape//2:, shape//2:]
    P1 = strassen(A11 + A22,B11 + B22)  
    P2 = strassen(A21 + A22,B11)        
    P3 = strassen(A11,B12 - B22)        
    P4 = strassen(A22,B21 - B11)        
    P5 = strassen(A11 + A12,B22)        
    P6 = strassen(A21 - A11,B11 + B12)  
    P7 = strassen(A12 - A22,B21 + B22)  
    C11 = P1 + P4 - P5 + P7  
    C12 = P3 + P5            
    C21 = P2 + P4            
    C22 = P1 - P2 + P3 + P6  
    C = np.vstack((np.hstack((C11, C12)), np.hstack((C21, C22))))
    return C
````

````python
def matrix(N):
    A = np.random.randint(0, 5, size=(N, N))
    B = np.random.randint(0, 5, size=(N, N))
    return A,B

def reshape(A,B):
    shapes = [2,4,8,16,32,64,128,256,512,1024]
    while A.shape[0] not in shapes:
            zeros = np.zeros(A.shape[0])
            A = np.vstack((A,zeros))
            B = np.vstack((B,zeros))
            zeros1 = np.zeros(A.shape[0]).reshape(A.shape[0],1)
            A = np.hstack((A,zeros1))
            B = np.hstack((B,zeros1))
    else:
        pass
    return A,B
````

````python
N_list = np.arange(100,800,100)
time_list_naive = np.zeros(len(N_list))
time_list_strassen = np.zeros(len(N_list))
time_list_built_in = np.zeros(len(N_list))
i = 0
for item in N_list:
    print(i)
    A,B = matrix(item)
    time1 = time.time()
    C = naive_multiplication(A,B)
    time2 = time.time()
    C = A @ B
    time3 = time.time()
    A,B = reshape(A,B)
    C = strassen(A,B)
    time4 = time.time()
    time_list_naive[i] = time2 - time1
    time_list_built_in[i] = time3 - time2
    time_list_strassen[i] = time4 - time3
    i += 1
    
#be patient, it's just code, but works very slow...
    

````

Naive multiplication has complexity $O(n^3)$, needed $2n^3$ operations for multiplication, Strassen decreases complexity to $O(n^{log_2(7)})$ by decreasing amount of values counted recursively by 1, but the main point is also that fact, that we need to rebuild matrix to 2^N, that also causes a huge impact in complexity. So for matrixes with $N = 2^n$ we have a victory in given range of N's, but when a huge amount of rows and columns needed to be added (like for N = 513 as example), naive started winning.

_2._ (3 pts) Complete a function ```compress```, that performs SVD and truncates it (using $k$ singular values/vectors). See the prototype below. 

Note, that in colourful case you have to split your image to channels and work with matrices corresponding to different channels separately.

Plot approximate reconstructed image $M_\varepsilon$ of your favorite image such that $rank(M_\varepsilon) = 5, 20, 50$ using ```plt.subplots```.

````python
def compress(image, k):
    """
    Perform svd decomposition and truncate it (using k singular values/vectors)
    
    Parameters: 
        image (np.array):        input image (probably, colourful)
        
        k (int):                 approximation rank
        
    --------
    Returns:
      reconst_matrix (np.array): reconstructed matrix (tensor in colourful case)
      
      s (np.array):              array of singular values 
    """
#     Your code is here
    # original shape(m;n) -> k(m+n) + k
    u,s,v = np.linalg.svd(image)
    u_k = u[:, :k]
    s_k = s[:k]
    v_k = v[:k, :]
    reconst_matrix = u_k.dot(np.diag(s_k).dot(v_k))
    return reconst_matrix, s
````

````python
# Your code is here 
def approximate(image,k):
    #colourfull picture 3-rd dimention (RGB) reshapes into the second dimention, then after compessing, reshapes back.
    n = 255
    image = image/n
    shape = image.shape
    reshape = image.reshape((shape[0],shape[1]*3))
    approx,s = compress(reshape,k)
    approx = approx.reshape(shape)
    plt.title('k={:d}'.format(k), fontsize= 20)
    plt.imshow(approx)
    return approx
````

````python
pic_raw = Image.open(requests.get(url1, stream=True).raw)
pic = np.array(pic_raw).astype(np.uint8)

plt.figure(figsize = (30,6))
plt.subplot(1, 5, 1)
approximate(pic,0)
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
plt.subplot(1, 5, 2)
approximate(pic,5)
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
plt.subplot(1, 5, 3)
approximate(pic,20)
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
plt.subplot(1, 5, 4)
approximate(pic,50)
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
plt.subplot(1,5,5)
plt.title('Original picture', fontsize = 20)
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
plt.imshow(pic_raw)
````

_3._ (3 pts) Plot the following two figures for your favorite picture
* How relative error of approximation depends on the rank of approximation?
* How compression rate in terms of storing information ((singular vectors + singular numbers) / total size of image) depends on the rank of approximation?

````python
# Your code is here

#1
def approx(image,k):
    n = 255
    image = image/n
    shape = image.shape
    reshape = image.reshape((shape[0],shape[1]*3))
    approx,s = compress(reshape,k)
    approx = approx.reshape(shape)
    compr = 100.0*(k*(shape[0] + shape[1])+k)/(shape[0]*shape[1])
    return approx, compr

def graph(pic):
    a = np.arange(1,100,5)
    er  = []
    comp = []
    for item in a:
        app,compr = approx(pic,item)
        er.append(np.linalg.norm(pic - app)/np.linalg.norm(pic))
        comp.append(compr)
    plt.figure(figsize=(15,7))
    plt.plot(a,er,lw = 2, c = 'r')
    plt.ylabel('Relative error', fontsize = 15)
    plt.xlabel('Rank of approximation', fontsize = 15)
    plt.title('Dependency of Relative error on Rank of approximation', fontsize = 20)
    plt.grid(True)
    plt.show()
    plt.figure(figsize=(15,7))
    plt.plot(a,comp, lw = 2, c ='b')
    plt.ylabel('Storing information', fontsize = 15)
    plt.xlabel('Rank of approximation', fontsize = 15)
    plt.title('Dependency of Storing information on Rank of approximation', fontsize = 20)
    plt.grid(True)
    plt.show()
    
    return 0
````

#1.The example is: $\|x\| = |x_1| + \sum_{i = 2}^n |x_i - x_1|$ is non absolute norm.

Let's prove that's a norm by definition:

1. $x = (0, \dots, 0) \rightarrow \|x\| = 0$
2.  $\forall x,y \in R^n \rightarrow \|x+y\| \leq \|x\| + \|y\| $
    $\|x+y\| = |x_1 + y_1| + \sum_{i = 2}^n |x_i + y_i - x_1 - y_1| \leq |x_1|  + \sum_{i = 2}^n |x_i - x_1| + + |y_1| \sum_{i = 2}^n |y_i - y_1| = \|x\| + \|y\|$
3. $\forall \alpha \in R, \forall x \in R^n \rightarrow \|\alpha x\| = |\alpha|\|x\|$:
    $\|\alpha x\| = |\alpha x_1| +\sum_{i = 2}^n |\alpha||x_i - x_1|  = |\alpha|\|x\| = |x_1| + \sum_{i = 2}^n |x_i - x_1| = |\alpha|\|x\|$
4. $\|x\| \geq 0$

So we proved, that it's a norm.
Concider $x = (1, -1, \dots, -1,1)$, then 
$\|x\| = 1+ 2(n-1)$, but $\||x| \| = 1$, then if $n > 2$ we receive non-absolute norm.
